Поиск загаданного числа.

Персонаж 1 загадал число от 1 до 128. Пусть x = 127.   
Персонаж 2 пытается угадать это число и задает вопросы, на которые есть ответ Да или Нет

### Тактика 1
1? Нет  
2? Нет  
3? Нет  
...  
127? Да

Какая сложность этого алгоритма?  
Сейчас мы находим одно число из 128, поэтому сложность будет не более чем 128 вопросов   
В лучшем случае сложность будет 1 - если мы загадали единицу.  
Можно посчитать, что в среднем нам потребуется 64 операций.  
Если угадываем от 1 до 1 000 000, то придется зададать до миллиона вопросов.  
В общем случае загадано число от 1 до n, то оно будет угадываться за O(n) раз.  

### Тактика 2
x > 64? Да  
x > 96? Да  
x > 112? Да  
x > 120? Да  
x > 124? Да  
x > 126? Да  
x > 127? Нет  
Тогда это 127!
Какая сложность этого алгоритма?  
Здесь нам потребовалось 7 вопросов, но что о большом массиве?  
Каждый раз мы мысленно делим наш массив на 2 части, потом еще 2 раза, потом еще в 2 раза.    
В итоге 1 единственный элемент мы получаем за O(log n) запросов.
