### Проверка числа на простоту 

Число является простым, если оно делится лишь на 1 и на само себя  
Как можно проверить, что число является простым?

1. Можно просто проверить, делится число на все между 2 и n - 1. O(n) 
2. Можно проверить, делится ли на 2, после чего проверить делимость на все нечетные числа от 3 до n - 1. Сложность будет также O(n), однако, константа перед n при подсчете будет в 2 раза меньше, если мы будем считать точное число операций.
3. Если число делится на 2 без остатка, то нам не нужно проверять, делится число на n / 2. А если число НЕ делится на 2, то оно точно не делится на n / 2.  
Если число не делится на 3, то оно не делится и на n/3.  
Если не делится на 5, то не делится и на n/5.
Если в целом (n / a) * a = n. Очевидно, что a <= n / a. Когда достигается равенство? Когда a = n / a. Т.е. a ** 2 = n.  
Поэтому нам максимальное число, на которое нужно проверять - это деление на корень из n, т.е. n ** 0.5.  
Сложность O(n ** 0.5)


Таким образом нужно проверять деление на 2, а так же на нечетные от 2 до n ** 0.5

### Решето Эратосфена
https://ru.wikipedia.org/wiki/Решето_Эратосфена

Задача: найти все простые числа от 1 до n.
1. В лоб - для каждого числа проверяем его на простоту. Всего n чисел, каждая проверка - это корень из n.  
n * n ** 0.5. Сложность О(n ** 1.5)
2. Решето Эратосфена. Мы вычеркиваем из списка чисел все, которые делятся на 2, потом на 3. 4 будет вычеркнуто. Вычеркиваем все, которые делятся на 5. Если число не вычеркнуто - оно простое.  

Тут нужно показать гифку из википедии

Алгоритм
1. Создадим массив из нулей
2. Просмастриваем массив. 
3. Если в очередной ячейке хранится 0, значит оно простое.
4. Если число простое, то нужно пройтись до конца массива и вместо 0 поставить 1 или другой признак. 

В конце из массива с помощью одного прохода можно получить все простые или все составные числа.
