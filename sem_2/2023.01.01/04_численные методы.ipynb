```python
import matplotlib.pyplot as plt
```

```python
def x_2(x):
    return x ** 2
```

```python
x = []
y = []
for i in range(-1000, 1000):
    x.append(i)
    y.append(x_2(i))
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(x, y);  # Plot some data on the axes.
```

```python
result = 0
# 0 до 10 
h = 0.0000001 # шаг
start, end = 0, 10
while start < end:
    result += h * x_2(start)
    start += h
```
